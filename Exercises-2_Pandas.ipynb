```python
cast = pd.read_csv('D:/TopMentor_DS_course/Advance_python_Pandas_S2/cast/cast.csv')
cast.head()
```

### What are the ten most common movie names of all time?

```python
cast['title'].value_counts().head(10)
```

### Which three years of the 1930s saw the most films released?

```python
cast[(cast['year']//10*10)==1930].groupby('year')['title'].nunique().sort_values(ascending=False).head(3)
```

### Plot the number of films that have been released each decade over the history of cinema.

```python
cast['decade'] = (cast['year']//10*10)
```

```python
cast.groupby('decade')['title'].nunique().reset_index(name='no_of_films')
```

### Plot the number of "Hamlet" films made each decade.

```python
cast[cast['title']=='Hamlet'].groupby('decade')['title'].size().reset_index(name='no_of_Hamlet_films')
```

### Plot the number of "Rustler" characters in each decade of the history of film.

```python
cast[cast['character']=='Rustler'].groupby('decade')['character'].size().reset_index(name='no_of_Rustler_character')
```

### Plot the number of "Hamlet" characters each decade.

```python
cast[cast['character']=='Hamlet'].groupby('decade')['character'].size().reset_index(name='no_of_Hamlet_character')
```
